```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/Week1_GA_dataset.csv')
print(df.head())
```

```python
#1: Count unknown ("?") values and replace with NaN
unknown_count = (df == '?').sum().sum()
print(f"Number of unknown ('?') values in the dataset: {unknown_count}")

df = df.replace('?', np.nan)
print("\nUnknown values replaced with NaN")
```

```python
#2: What is the shape of the dataset?
shape = df.shape
print(f"Shape of the dataset: {shape}")
print(f"Answer: ({shape[0]}, {shape[1]})")
```

```python
#3: Value at 692th indexed row and 0th indexed column
value_692_0 = df.iloc[692, 0]
print(f"Value at df[692, 0]: {value_692_0}")
```

```python
#4: Value at 546th indexed row and 7th indexed column
value_546_7 = df.iloc[546, 7]
print(f"Value at df[546, 7]: {value_546_7}")
```

```python
#5: Unique values in Locality feature
unique_locality = df['Locality'].unique()
print("Unique values in Locality feature:")
print(sorted([str(x) for x in unique_locality if pd.notna(x)]))
```

```python
#6: Features with missing (NaN) values
missing_features = df.columns[df.isnull().any()].tolist()
print("Features with missing (NaN) values:")
for feature in missing_features:
    print(f"  - {feature}: {df[feature].isnull().sum()} missing values")
```

```python
#7: Feature with most missing (NaN) values
missing_counts = df.isnull().sum()
most_missing_feature = missing_counts.idxmax()
most_missing_count = missing_counts.max()
print(f"Feature with most missing values: {most_missing_feature}")
print(f"Number of missing values: {most_missing_count}")
```

```python
#8:Drop rows with missing values strictly greater than 2
df_dropped_gt2 = df[df.isnull().sum(axis=1) <= 2]
samples_remaining = len(df_dropped_gt2)
print(f"Samples remaining after dropping rows with missing values > 2: {samples_remaining}")
```

```python
#9: Drop all rows with any missing values
df_no_missing = df.dropna()
samples_remaining_no_na = len(df_no_missing)
print(f"Samples remaining after dropping all rows with missing values: {samples_remaining_no_na}")
```
