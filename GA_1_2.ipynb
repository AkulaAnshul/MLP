```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Week1_GA_dataset.csv', na_values=['?'])

print("="*80)
print("DATASET INFO")
print("="*80)
print(f"Shape of dataset: {df.shape}")
print("\nFirst few rows:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())

print("\n" + "="*80)
print("QUESTION 1: Even indexed rows - value at [0, 3]")
print("="*80)
df_even_rows = df.iloc[::2]
value_even_0_3 = df_even_rows.iloc[0, 3]
print(f"Answer: {value_even_0_3}")
print(f"Column name: {df.columns[3]}")

print("\n" + "="*80)
print("QUESTION 2: Even indexed rows - value at [332, 3]")
print("="*80)
df_even_rows = df.iloc[::2]
value_even_332_3 = df_even_rows.iloc[332, 3]
print(f"Answer: {value_even_332_3}")

print("\n" + "="*80)
print("QUESTION 3: Odd indexed rows - value at [332, 3]")
print("="*80)
df_odd_rows = df.iloc[1::2]
value_odd_332_3 = df_odd_rows.iloc[332, 3]
print(f"Answer: {value_odd_332_3}")
print(f"Type: {type(value_odd_332_3)}")

print("\n" + "="*80)
print("QUESTION 4: Odd indexed columns - value at [100, 5]")
print("="*80)
df_odd_cols = df.iloc[:, 1::2]
print(f"Odd indexed column names: {df_odd_cols.columns.tolist()}")
value_odd_cols_100_5 = df_odd_cols.iloc[100, 5]
print(f"Answer: {value_odd_cols_100_5}")

print("\n" + "="*80)
print("QUESTION 5: Even indexed columns and rows - value at [255, 3]")
print("="*80)
df_even_both = df.iloc[::2, ::2]
print(f"Even indexed column names: {df_even_both.columns.tolist()}")
value_even_both_255_3 = df_even_both.iloc[255, 3]
print(f"Answer: {value_even_both_255_3}")

print("\n" + "="*80)
print("QUESTION 6: Samples from six most recent years")
print("="*80)
six_most_recent_years = sorted(df['Year'].dropna().unique(), reverse=True)[:6]
print(f"Six most recent years: {six_most_recent_years}")
samples_recent_years = df[df['Year'].isin(six_most_recent_years)].shape[0]
print(f"Answer: {samples_recent_years}")

print("\n" + "="*80)
print("QUESTION 7: Samples with num_rooms = 3 AND num_bathrooms = 3")
print("="*80)
samples_3_and_3 = df[(df['num_rooms'] == 3) & (df['num_bathrooms'] == 3)].shape[0]
print(f"Answer: {samples_3_and_3}")

print("\n" + "="*80)
print("QUESTION 8: Samples with num_rooms = 3 OR num_bathrooms = 3")
print("="*80)
samples_3_or_3 = df[(df['num_rooms'] == 3) | (df['num_bathrooms'] == 3)].shape[0]
print(f"Answer: {samples_3_or_3}")

print("\n" + "="*80)
print("QUESTION 9: Year 2022, Greenwich, num_rooms = 3, facing North or East")
print("="*80)
samples_2022_greenwich = df[
    (df['Year'] == 2022) &
    (df['Locality'] == 'Greenwich') &
    (df['num_rooms'] == 3) &
    ((df['Face'] == 'North') | (df['Face'] == 'East'))
].shape[0]
print(f"Answer: {samples_2022_greenwich}")

print("\n" + "="*80)
print("QUESTION 10: Samples from August")
print("="*80)
df['Date_parsed'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date_parsed'].dt.month
samples_august = df[df['Month'] == 8].shape[0]
print(f"Answer: {samples_august}")

print("\n" + "="*80)
print("QUESTION 11: Locality with highest average Sale Price")
print("="*80)
avg_sale_price_by_locality = df.groupby('Locality')['Sale Price'].mean().sort_values(ascending=False)
print("Average Sale Price by Locality:")
print(avg_sale_price_by_locality)
print(f"\nAnswer: {avg_sale_price_by_locality.idxmax()}")

print("\n" + "="*80)
print("ALL ANSWERS COMPLETE")
print("="*80)
```
